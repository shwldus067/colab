```
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
cancer=load_breast_cancer()
x=cancer.data
y=cancer.target

x_train_all, x_test, y_train_all, y_test=train_test_split(x, y, stratify=y, 
                                                test_size=0.2, random_state=42)
x_train, x_val, y_train, y_val=train_test_split(x_train_all, y_train_all,
                          stratify=y_train_all, test_size=0.2, random_state=42)

#표준화
train_mean=np.mean(x_train, axis=0)
train_std=np.std(x_train, axis=0)
x_train_scaled=(x_train-train_mean)/train_std
x_val_scaled=(x_val-train_mean)/train_std;

sgd=SGDClassifier(loss='hinge', random_state=42)
sgd.fit(x_train, y_train)
sgd.score(x_val, y_val)
```

```
w=[1, 2, 3, -4, 0]  #np.sign() : 값의 부호를 반환
print(np.sign(w))
```

```
class SingleLayer():
  def __init__(self, learning_rate=0.1, l1=0, l2=0):
    self.w=None   #데이터 특성이 많기 때문에 미리 초기화하지 않음
    self.b=None
    self.losses=[]   #훈련 세트 손실 변화 저장
    self.w_history=[] #가중치의 변화 저장
    self.val_losses=[]  #검증 세트의 손실 변화 저장
    self.lr=learning_rate
    self.l1=l1    #l1, l2 규제의 강도를 조절하는 매개변수
    self.l2=l2

  def forpass(self, x):
    z=np.sum(x*self.w)+self.b
    return z
  
  def backprop(self, x, err):
    w_grad=x*err
    b_grad=1*err
    return w_grad, b_grad

  def fit(self, x, y, epochs=100, x_val=None, y_val=None):
    self.w=np.ones(x.shape[1])
    self.b=0
    self.w_history.append(self.w.copy())  #가중치 기록
    np.random.seed(42)  #시드갑 지정
    for i in range(epochs):
      loss=0
      indexes=np.random.permutation(np.arange(len(x))) #인덱스를 섞어 훈련데이터 섞음
      for i in indexes:
        z=self.forpass(x[i])  #정방향 계산
        a=self.activation(z)  #활성화 함수 적용
        err=-(y[i]-a) #오차 계산
        w_grad, b_grad=self.backprop(x[i], err) #역방향 계산
        #그래디언트에 패널티 항의 미분값을 더함
        w_grad+=self.l1*np.sign(self.w)+self.l2*self.w
        self.w-=w_grad*self.lr  #학습률 적용
        self.b-=b_grad
        self.w_history.append(self.w.copy())  #가중치 기록
        a=np.clip(a, 1e-10, 1-1e-10)  #안전한 로그 계산을 위한 클리핑
        loss+=-(y[i]*np.log(a)+(1-y[i])*np.log(1-a))  #에포크마다 손실 합 계산
      self.losses.append(loss/len(y)+self.reg_loss()) #에포크마다 평균 손실 저장
      self.update_val_loss(x_val, y_val)
  
  def reg_loss(self):
    return self.l1*np.sum(np.abs(self.w))+self.l2/2*np.sum(self.w**2)

  def update_val_loss(self, x_val, y_val):
    if x_val is None: #검증 데이터가 입력됐을 때만 수행
      return
    val_loss=0
    for i in range(len(x_val)): #검증 데이터를 하나씩 계산
      z=self.forpass(x_val[i])
      a=self.activation(z)
      a=np.clip(a, 1e-10, 1-1e-10)
      val_loss+=-(y_val[i]*np.log(a)+(1-y_val[i])*np.log(1-a))  #손실값의 합
    self.val_losses.append(val_loss/len(y_val)+self.reg_loss()) #손실 기록
  
  def activation(self, z):
    a=1/(1+np.exp(-z))
    return a

  def predict(self, x):
    z=[self.forpass(x_i) for x_i in x]  #hypothesis를 구함
    return np.array(z)>0  #스텝 함수 적용

  def score(self, x, y):
    return np.mean(self.predict(x)==y)
```

```
#l1 규제
l1_list=[0.0001, 0.001, 0.01]
for l1 in l1_list:
  lyr=SingleLayer(l1=l1)
  lyr.fit(x_train_scaled, y_train, x_val=x_val_scaled, y_val=y_val)
  #훈련 세트와 검증 세트의 에포크에 따른 손실 그래프
  plt.plot(lyr.losses)
  plt.plot(lyr.val_losses)
  plt.title('Lerning Curve (l1={})'.format(l1))
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['train_loss', 'val_loss'])
  plt.ylim(0, 0.3)
  plt.show()
  #가중치들의 분포 그래프
  plt.plot(lyr.w, 'bo')
  plt.title('Weight (l1={})'.format(l1))
  plt.ylabel('value')
  plt.xlabel('weight')
  plt.ylim(-4, 4)
  plt.show()
```

```
layer=SingleLayer(l1=0.001) #0.01은 과소적합이 생기므로 0.001로 규제
#l1=0.001 그래프상의 최적점인 epoch=20값을 준다.
layer.fit(x_train_scaled, y_train, epochs=20)
layer.score(x_val_scaled, y_val)
```

```
#l2 규제
l2_list=[0.0001, 0.001, 0.01]
for l2 in l2_list:
  lyr=SingleLayer(l2=l2)
  lyr.fit(x_train_scaled, y_train, x_val=x_val_scaled, y_val=y_val)
  #훈련 세트와 검증 세트의 에포크에 따른 손실 그래프
  plt.plot(lyr.losses)
  plt.plot(lyr.val_losses)
  plt.title('Learning Curve (l2={})'.format(l2))
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(['train_loss', 'val_loss'])
  plt.ylim(0, 0.3)
  plt.show()
  #가중치들의 분포 그래프
  plt.plot(lyr.w, 'bo')
  plt.title('Weight (l2={})'.format(l2))
  plt.ylabel('value')
  plt.xlabel('weight')
  plt.ylim(-4, 4)
  plt.show()
```

```
layer2=SingleLayer(l2=0.01) #적당한 l2규제 강도는 0.01이다.
layer2.fit(x_train_scaled, y_train)
layer2.score(x_val_scaled, y_val)
```

```
#SGD class사용한 구현
sgd=SGDClassifier(loss='log', penalty='l2', alpha=0.001, random_state=42)
sgd.fit(x_train_scaled, y_train)
sgd.score(x_val_scaled, y_val)
```

```
validation_scores=[]
k=10    #나눌 fold 갯수
bins=len(x_train)//k #fold 하나의 길이

for i in range(k):
  start=i*bins
  end=(i+1)*bins
  val_fold=x_train_all[start:end] #데이터 폴드
  val_target=y_train_all[start:end] #타깃 폴드

  #검증 세트 폴드 제외한 훈련 데이터 폴드 인덱스
  train_index=list(range(0, start))+list(range(end, len(x_train)))
  train_fold=x_train_all[train_index]
  train_target=y_train_all[train_index]

  train_mean=np.mean(train_fold, axis=0)
  train_std=np.std(train_fold, axis=0)
  train_fold_scaled=(train_fold-train_mean)/train_std
  val_fold_scaled=(val_fold-train_mean)/train_std

  lyr=SingleLayer(l2=0.01)
  layer.fit(x_train_scaled, y_train, epochs=20)
  lyr.fit(train_fold_scaled, train_target, epochs=50)
  score=lyr.score(val_fold_scaled, val_target)
  validation_scores.append(score)

print(np.mean(validation_scores))
```

```
from sklearn.model_selection import cross_validate  #사이킷런의 교차검증 모델
sgd=SGDClassifier(loss='log', penalty='l2', alpha=0.001, random_state=42)
scores=cross_validate(sgd, x_train_all, y_train_all, cv=10)
print(np.mean(scores['test_score']))  #전천리를 하지 않아 정확도가 낮음
```

```
from sklearn.pipeline import make_pipeline  #전처리와 모델을 합치기 위한 파이프라인
from sklearn.preprocessing import StandardScaler   #전처리를 위한 모듈
pipe=make_pipeline(StandardScaler(), sgd)
scores=cross_validate(pipe, x_train_all, y_train_all, cv=10, return_train_score=True)
print(np.mean(scores['test_score']))  #교차 검증 정확도
print(np.mean(scores['train_score'])) #훈련 데이터의 정확도
```
